## Simulating Central Limit Theorem  (CLT)
Let's simulate an experiment:
1. r.v X = average(Roll an n sided dice N times)
2. Repeat above experiment K times
3. CLT states that the X tends to be a normal for large K, regardless of values of n and N, however as observed below, larger values of n and N can help bring convergence earlier

```python
## Simulating a dice roll sum 
K_vals = [100, 10000]
N_vals = [100, 10000]
width = len(K_vals)
height = len(N_vals)
fig, ax = plt.subplots(height, width, sharex='col', sharey='row', figsize=(15, 15))
for i, K in enumerate(K_vals):
    for j, N in enumerate(N_vals):
        X_arr = [] ## to store results of experiment
        for k in range(K): #Perform experiment K times
            X_arr.append(np.mean(np.random.randint(1, 7, N)))
#             ax[i-1, j-1].text(0.5, 0.5, str((i, j)), fontsize=18, ha='center')
        sns.distplot(a=X_arr, ax=ax[i, j], label="N={}, K={}".format(N, K))
        ax[i, j].legend()
#       sns.distplot(X_arr)
```
